```python
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
# load dataset

dataframe = pd.read_csv("LatestCases2020.csv")
data = dataframe.values
# split into inputs and outputs
X, y = data[:, :-2], data[:, -2]
print(X.shape, y.shape)
# split into train test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
# fit the model
model = RandomForestClassifier(random_state=1)
model.fit(X_train, y_train)
# make predictions
yhat = model.predict(X_test)
# evaluate predictions
acc = accuracy_score(y_test, yhat)
print('Accuracy: %.3f' % acc)
```

```python
import pandas as pd
from sklearn import linear_model
# load the dataset into memory
training_dataset = pd.read_csv("LatestCases2020.csv")
# create a model using the linear regression algorithm
# and train it with the data from our csv
regression_model = linear_model.LinearRegression()
print ("Training model...")
# model training
regression_model.fit(training_dataset[['Age Group']], training_dataset.CurrentlyHospitalized) 
print ("Model trained.")
# ask user to enter an area and calculate
# its price using our model
input_area = int(input("Enter Age Group: "))
currently_hospitalized = regression_model.predict([[input_area]])
print ("Currently Hospitalized:", round(currently_hospitalized[0], 2))
```

```python
import pandas
from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
#from IPython.display import Image

df = pandas.read_csv("LatestCases2020.csv")

print(df)
```

```python

features = ['_id', 'Age Group']

X = df[features]
y = df['CurrentlyHospitalized']

print(X)
print(y)
```

```python
dtree = DecisionTreeClassifier()
data = dtree.fit(X, y)#sample_weight=None, check_input=True, X_idx_sorted=None
data = tree.export_graphviz(dtree, out_file=None, feature_names=features)
graph = pydotplus.graph_from_dot_data(data)
graph.write_png('mydecisiontree.png')
img=pltimg.imread('mydecisiontree.png')
plt.figure(figsize = (14, 18))
imgplot = plt.imshow(img)
#Image(filename = "mydecisiontree.png", width = 600, height = 300)
plt.show()
```
